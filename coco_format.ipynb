```python
import json
import pandas as pd
from pathlib import Path
import os

json_file = Path(r'table\annotations\instances_default.json')

with open(json_file, 'r') as f:
    Json = json.load(f)
images = pd.DataFrame(Json['images'])
file_name = images['file_name'].apply(lambda x : f'{x.split('_')[0]}_{x.split('_')[1]}.jpg')
images['file_name'] = file_name
categories = pd.DataFrame(Json['categories'])
annotations = pd.DataFrame(Json['annotations'])
```

```python
with open(r'BP-CBI\annotations\instances_default.json', 'r') as f:
    myJson = json.load(f)
myJson['images']
```

```python
BRANCH_FILTER = 'JA-CBI'
file_name_list = [val for val in images['file_name'].values  if val.split('_')[0] == BRANCH_FILTER]
file_name_list
```

```python
filter_img = images[images['file_name'].isin(file_name_list)]
filter_img
```

```python
import numpy as np
anno = []
img = []
count = 0
for im in filter_img.iloc:
    anno_df = annotations.query('`image_id` == @im.id & `category_id` == 3')
    img.append({
        'id': int(im.id),
        'width': int(im.width),
        'height': int(im.height),
        'file_name': im.file_name,
        'license': int(0),
        'flickr_url': '',
        'coco_url': '',
        'date_captured': int(0)
    })
    for an in anno_df.iloc:
        count += 1
        anno.append({
            "id" : int(count),
            "image_id": int(an.image_id),
            "category_id": int(1),
            "segmentation": an.segmentation,
            "area": float(an.area),
            "bbox": an.bbox,
            "iscrowd": 0,
            "attributes": {
                "occluded": False
            }
        })
        count += 1
        anno.append({
            "id" : int(count),
            "image_id": int(an.image_id),
            "category_id": int(2),
            "segmentation": an.segmentation,
            "area": float(an.area),
            "bbox": an.bbox,
            "iscrowd": 0,
            "attributes": {
                "occluded": False
            }
        })
    # break
```

```python
# COCO json format
coco_format = {
    "licenses": myJson['licenses'][0],
    "info":myJson['info'],
    "categories": myJson['categories'],
    "images": img,
    "annotations": anno
}
```

```python
with open('coco_format.json', 'w') as json_file:
    json.dump(coco_format, json_file, indent=5)
```
